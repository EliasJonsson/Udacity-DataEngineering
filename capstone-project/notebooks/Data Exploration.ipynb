### 2. Extract Unique values


```python
def extract_info(df, n):
    countLabel = "Frequency"
    print(f"Number of values for this table is {df.shape[0]}")
    for col in df:
        df[countLabel] = df.groupby(col)[col].transform('count')
        df.sort_values(countLabel, inplace=True, ascending=False)
        counter = df[[col, countLabel]]
        counter = counter.drop_duplicates()
        print(f"Column: {col}")
        print(f"Total number of unique values are {len(df[col].unique())}")
        print(f"{n} most frequent values")
        print(f"{counter.head(n)}")
        print(f"Number of NaN values: {df[col].isna().sum()}, {df[col].isna().sum()/df.shape[0] * 100}%)")
        print("______________________")
    df.drop(columns=[countLabel], inplace=True)
```
